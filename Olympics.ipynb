```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
athletes = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/athlete_events.csv')
athletes.describe()
```

```python
cleaned_athletes = athletes.info()
```

```python
athletes[['Name', 'Height']]
```

```python
above_25_filter = athletes['Age'] > 25
athletes[above_25_filter]
```

```python
non_null_height = athletes['Height'].notnull()
non_null_weight = athletes['Weight'].notnull()
non_null_height_and_weight_filter = non_null_height & non_null_weight
athletes[non_null_height_and_weight_filter]
```

```python
# What is the median height/weight of an Olympic athlete?

athletes[non_null_height_and_weight_filter][['Height', 'Weight']].median()
```

```python
# What is the median height/weight of an Olympic medalist?

# What is the median height/weight of an Olympic medalist?
is_medalist = athletes['Medal'].notnull()
athletes[is_medalist & non_null_height_and_weight_filter][['Height', 'Weight']].median()
```

```python
plt.figure(figsize=(10,10))
plt.title("Height and Weight distribution of Olympics Athlets")
sns.scatterplot(data=athletes, x="Height", y="Weight")
plt.show()
```

2. What is the best age to win a golden medal?
Answer 22


```python
# prompt: What is the best age to win a golden medal?

is_gold_medalist = athletes['Medal'] == 'Gold'
gold_medalists = athletes[is_gold_medalist & non_null_height_and_weight_filter]
gold_medalists['Age'].value_counts().idxmax()
```

```python
non_null_age = athletes['Age'].notnull()
non_null_age_df = athletes[non_null_age]
non_null_age_df.head()
```

```python
golden_medalists_filter = non_null_age_df['Medal'] == 'Gold'
golden_medalists_df = non_null_age_df[golden_medalists_filter]
golden_medalists_df.head()
```

```python
# @title Age

from matplotlib import pyplot as plt
golden_medalists_df['Age'].plot(kind='hist', bins=40, title='Age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Average Age of Gold Medalists Over Time

import matplotlib.pyplot as plt

golden_medalists_df.groupby(['Year', 'Sex'])['Age'].mean().unstack().plot()
plt.xlabel('Year')
plt.ylabel('Average Age')
_ = plt.title('Average Age of Gold Medalists Over Time')
```

```python
# @title Number of Gold Medals Awarded Over Time

import matplotlib.pyplot as plt

golden_medalists_df.groupby(['Year', 'Season'])['Medal'].count().unstack().plot()
plt.xlabel('Year')
plt.ylabel('Number of Gold Medals')
_ = plt.title('Number of Gold Medals Awarded Over Time')
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(data=golden_medalists_df, x='Sport', palette='magma', hue='Age')
plt.xticks(rotation=90)
plt.show()
```

3. Which countries have won the most medals? Overall and per year?


```python
# Which countries have won the most medals? Overall and per year?

# Overall
athletes.groupby('NOC')['Medal'].count().sort_values(ascending=False).head()

# Per Year
athletes.groupby(['Year', 'NOC'])['Medal'].count().sort_values(ascending=False).head()
```

```python
# Which year the most competitive Olympics?
athletes.groupby('Year')['Name'].nunique().sort_values(ascending=False).head()
```
